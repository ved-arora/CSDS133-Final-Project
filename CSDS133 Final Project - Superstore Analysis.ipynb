```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

## Importing / Understanding the data

```python
#Calling the entire dataset (uncleaned) "bigdata"
#bigdata = pd.read_csv('Sample - Superstore.csv', encoding='cp1252')
bigdata = pd.read_csv('Sample - Superstore.csv', encoding='windows-1252')
bigdata
```

```python
bigdata.head()
```

```python
#understanding the data types of each column
bigdata.info()
```

```python
#Understanding how which countries are represented in this data
#The output tells us that only the USA is represented, so we know we can drop this column full of just 'United States'
bigdata.Country.unique()
```

## Cleaning the data

```python
#we know that we can get rid of the 'Country' column because for every row, it is filled with just the USA
bigdata.drop('Country', axis = 1, inplace = True)
#we know that we can get rid of the row ID because Jupyter automatically gives the data row IDs
bigdata.drop('Row ID', axis = 1, inplace = True)
#we know that we can get rid of both the customer ID and customer name because we are interested in the bigger picture and have no plans to index by name or customer ID
bigdata.drop('Customer ID', axis = 1, inplace = True)
bigdata.drop('Customer Name', axis = 1, inplace = True)
#replacing spaces with underscores from each of the column names (for ease of access)
bigdata.columns = bigdata.columns.str.replace(' ','_')
```

```python
bigdata
```

```python
#We want to know how many categories of products this company is selling, so we can split each category up later and perform more specific analysis
#The outcome shows us that there are only 3 main categories that we must analyze in depth
bigdata.Category.unique()
```

```python
#This is a test to see if the product ID will have the same count as the Product Name
#The result shows us that they are the same, meaning we can get rid of one of the columns to make the data a bit cleaner
x = (bigdata['Product_ID'] == 'FUR-BO-10001798').sum()
y = (bigdata['Product_Name'] == 'Bush Somerset Collection Bookcase').sum()
print(x)
print(y)
```

```python
#Creating a new dataframe called cleandata (indicating that now we are using the clean data)
cleandata = bigdata
#we will not be specifically indexing specific order IDs so we can get rid of this column as well
cleandata.drop('Order_ID', axis = 1, inplace = True)
cleandata
```

## Creating New Data Tables based on Category

```python
#This is the data for the furniture category
furnitureData = cleandata[cleandata['Category'] == 'Furniture']
furnitureData
```

```python
#This is the data for the Technology category
techData = cleandata[cleandata['Category'] == 'Technology']
techData
```

```python
#This is the data for the office supplies category
officeData = cleandata[cleandata['Category'] == 'Office Supplies']
officeData
```

## Data Visualization

```python
cleandata['Category'].value_counts().plot(kind='bar',title='Number of Orders for each Major Category')
plt.xticks(rotation=45, ha='right')
plt.xlabel("Category")
plt.ylabel("# of Orders")
plt.savefig('Orders vs Category BAR.png')
```

```python
cleandata.groupby('Category')['Profit'].sum().plot(kind='bar', title='Major Category Total Profit ($)', ylabel='Total Profit ($)')
plt.xticks(rotation=45, ha='right')
plt.savefig('Orders vs Profit BAR.png')
```

```python
from scipy import stats
fig, ax = plt.subplots(figsize=(9,6))
x = cleandata['Sales']
y = cleandata['Profit']
slope, intercept, r, p, std_err = stats.linregress(x, y)
def myfunc(x):
    return slope * x + intercept
mymodel = list(map(myfunc, x))
sctr = ax.scatter(x,y)
ax.plot(x, mymodel)
ax.set_title('Overall Profit vs. Sales', fontsize=14)
ax.set_xlabel('Sales ($)')
ax.set_ylabel('Profit ($)')
plt.savefig('BEFORE overall profit vs sales scatter.png')
print('y =',slope,'x',' + ',intercept)
print(r)
```

```python
import seaborn as sns
labels = cleandata.groupby('Region')['Profit'].sum().index
values = cleandata.groupby('Region')['Profit'].sum().values
colors = sns.color_palette('pastel')[0:4]
plt.pie(values, labels = labels, colors = colors, autopct='%.0f%%')
plt.title("Total Profit by Region")
plt.show()
plt.savefig('BEFORE region PIE.png')
```

## Data Analysis (Furniture)

```python
#understanding and showing how many different furniture products were sold
print(furnitureData.Product_ID.nunique())
```

```python
#creating a dataframe of orders that resulted in negative profit (losing money)
negativeFurnitureData = furnitureData[furnitureData['Profit'] < 0]
negativeFurnitureData
```

```python
#we know from the dataframe that there are 714 orders that resulted in negative profit
#we want to find out how many unique products there are out of these 714 orders
print(negativeFurnitureData.Product_ID.nunique())
```

```python
counts = negativeFurnitureData['Product_ID'].value_counts()
#finding the products that had 6 or more orders with negative profit
negativeFurnitureData = negativeFurnitureData[~negativeFurnitureData['Product_ID'].isin(counts[counts < 4].index)]
negativeFurnitureData
```

## Furniture Data Regressions (with and without bad suggested products removed)

```python
#plotting a scatterplot comparing sales (x axis) against profit (y axis)
fig, ax = plt.subplots(figsize=(9,6))
x = furnitureData['Sales']
y = furnitureData['Profit']
slope, intercept, r, p, std_err = stats.linregress(x, y)
def myfunc(x):
    return slope * x + intercept
mymodel = list(map(myfunc, x))
sctr = ax.scatter(x,y)
ax.plot(x, mymodel)
ax.set_title('Furniture Profit vs. Sales (for each order)', fontsize=14)
ax.set_xlabel('Sales ($)')
ax.set_ylabel('Profit ($)')
#printing our linear regression model
print('y =',slope,'x',' + ',intercept)
#printing our correlation coefficient
print(r)
plt.savefig('BEFORE FURN profit vs sales.png')
```

```python
#removing the items that had 4+ appearences on the negative profit data from the big Furniture dataframe
furnitureData = furnitureData[~furnitureData.Product_ID.isin(negativeFurnitureData.Product_ID)]
#creating a dataframe made of furniture that had anything equal to or below negative $500 profit (any loss of $500 or more)
worstFurniture = furnitureData[furnitureData['Profit'] < -500]
worstFurniture
```

```python
#removing the $500 or more loss items from the big dataframe
furnitureData = furnitureData[~furnitureData.Product_ID.isin(worstFurniture.Product_ID)]
#plotting a scatterplot comparing sales (x axis) against profit (y axis) of the new data (where our suggested products are removed)
fig, ax = plt.subplots(figsize=(9,6))
x = furnitureData['Sales']
y = furnitureData['Profit']
slope, intercept, r, p, std_err = stats.linregress(x, y)
def myfunc(x):
    return slope * x + intercept
mymodel = list(map(myfunc, x))
sctr = ax.scatter(x,y)
ax.plot(x, mymodel)
ax.set_title('Projected Furniture Profit vs. Sales (after suggested removals)', fontsize=14)
ax.set_xlabel('Sales ($)')
ax.set_ylabel('Profit ($)')
print('y =',slope,'x',' + ',intercept)
print(r)
plt.savefig('AFTER FURN profit vs sales.png')
```

```python
#creating a dataframe that will keep the name of every one of our suggested removals (so we can submit to the company)
frames = [negativeFurnitureData, worstFurniture]
removedFurnitureItems = pd.concat(frames)
removedFurnitureItems.drop(removedFurnitureItems.columns.difference(['Product_ID', 'Product_Name']), 1, inplace=True)
#removedFurnitureItems.to_csv('List_of_Removed_Furniture_Items.csv')
```
